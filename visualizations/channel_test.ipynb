```python
import nibabel as nib
import numpy as np
```

```python
nii_path = r"C:\Users\justi\OneDrive\Desktop\University\Master_Studium\MP\dataset\synth23_pelvis_v7_nifti_con\test\Ungood_whole_patient_scans\img\1PA025_26.nii"

img = nib.load(nii_path)
data = img.get_fdata()

print("Data shape:", data.shape)

# --- Compare channels ---
diff_01 = np.abs(data[:, :, 0, 0] - data[:, :, 0, 1]).mean()
diff_12 = np.abs(data[:, :, 0, 1] - data[:, :, 0, 2]).mean()

print("Mean absolute difference between ch0 and ch1:", diff_01)
print("Mean absolute difference between ch1 and ch2:", diff_12)

if diff_01 < 1e-6 and diff_12 < 1e-6:
    print("✅ Likely new 1-channel (repeated 3 times).")
else:
    print("🧩 Likely old 2.5D version (distinct slices).")
```
